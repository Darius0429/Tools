# ViT

## Configuration

```python
img_size = 224
kernel_size = 16
dim_tokens = 768
num_layers = 4
num_heads = 4

hidden_dim = dim_tokens * 4
dropout = 0.0
```

## Embedding

```python
import torch
from torch import nn
from nn.encoders.vit import ViTEncoder

dummy_input = torch.randn(1, 3, img_size, img_size)

encoder = ViTEncoder(
    num_channels=3,
    output_dim=dim_tokens,
    kernel_size = kernel_size,
)
dummy_encoded = encoder(dummy_input)
```

## Transformer layers

```python
from nn.transformers.basic_transformer import Transformer

transformer = Transformer(
    dim_tokens=dim_tokens,
    num_layers=num_layers,
    num_heads=num_heads,
    dropout=dropout,
    hidden_dim=hidden_dim,
)

output = transformer(dummy_encoded)
```
